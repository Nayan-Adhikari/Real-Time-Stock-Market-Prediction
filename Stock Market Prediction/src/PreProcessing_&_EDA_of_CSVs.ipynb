### Preprocessing of Real Time data

```python
import pandas as pd
```

```python
data_rl=pd.read_csv("Downloads/AAPL_real_time_data.csv")
```

```python
data_rl.head()
```

```python
data_rl['Datetime'] = pd.to_datetime(data_rl['Datetime'])
```

```python
data_rl.info()
```

```python
data_rl.head()
```

```python
data_rl.head()
```

```python
data_rl.head()
```

```python
data_rl.head()
```

```python
columns_order = ['Date', 'Time'] + [col for col in data_rl.columns if col not in ['Date', 'Time']]
data_rl = data_rl[columns_order]
```

```python
data_rl['Date','Time'] = pd.to_datetime(data_rl['Date','Time'])
data['Date'] = pd.to_datetime(data['Date'])
```

```python
print("Real-Time Data Info:")
print(data_rl.info())
print("\nHistorical Data Info:")
print(data.info())
```

```python
print("\nReal-Time Data Sample:")
print(data_rl.head())
print("\nHistorical Data Sample:")
print(data.head())
```

```python
print("\nMissing Values in Real-Time Data:")
print(data_rl.isnull().sum())
print("\nMissing Values in Historical Data:")
print(data.isnull().sum())
```

```python
print("\nSummary Statistics for Real-Time Data:")
print(data_rl.describe())
print("\nSummary Statistics for Historical Data:")
print(data.describe())
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(data_rl['Close'], bins=50, kde=True, color='blue', label='Real-Time Data')
sns.histplot(data['Close'], bins=50, kde=True, color='orange', label='Historical Data')
plt.title('Distribution of Closing Prices')
plt.xlabel('Closing Price')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.plot(data_rl['Date'], data_rl['Close'], label='Real-Time Data', color='blue')
plt.plot(data['Date'], data['Close'], label='Historical Data', color='orange')
plt.title('Closing Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend()
plt.show()
```

```python
# Filter numeric columns
numeric_columns_real_time = data_rl.select_dtypes(include=['float64', 'int64'])
numeric_columns_historical = data.select_dtypes(include=['float64', 'int64'])

# Correlation heatmap for Real-Time Data
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_columns_real_time.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap (Real-Time Data)')
plt.show()

# Correlation heatmap for Historical Data
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_columns_historical.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap (Historical Data)')
plt.show()
```

```python
# Calculate daily returns
data_rl['Daily_Return'] = data_rl['Close'].pct_change()
data['Daily_Return'] = data['Close'].pct_change()

# Plot daily returns
plt.figure(figsize=(14, 7))
plt.plot(data_rl['Date'], data_rl['Daily_Return'], label='Real-Time Daily Return', alpha=0.7)
plt.plot(data['Date'], data['Daily_Return'], label='Historical Daily Return', alpha=0.7)
plt.title('Daily Returns Over Time')
plt.xlabel('Date')
plt.ylabel('Daily Return')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

# Perform Augmented Dickey-Fuller test
def check_stationarity(data, column):
    result = adfuller(data[column].dropna())
    print(f"ADF Statistic for {column}: {result[0]}")
    print(f"p-value: {result[1]}")
    if result[1] <= 0.05:
        print("The data is stationary.")
    else:
        print("The data is not stationary.")

check_stationarity(data_rl, 'Close')
check_stationarity(data, 'Close')
```

```python
print("Real-Time Data Date Range:", data_rl['Date'].min(), "to", data_rl['Date'].max())
print("Historical Data Date Range:", data['Date'].min(), "to", data['Date'].max())
```
